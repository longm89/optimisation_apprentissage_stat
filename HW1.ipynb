```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
mat = scipy.io.loadmat("data_orsay_2017.mat")
Xtrain = mat["Xtrain"]
ytrain = mat["ytrain"]
Xtest = mat["Xtest"]
ytest = mat["ytest"]
```

```python
# For the square loss
    
def square_loss(y, theta, x):
    theta_dot_x = np.sum(theta * x)
    return (1 - float(y) * theta_dot_x)**2

def square_loss_derivative(y, theta, x):
    theta_dot_x = np.sum(theta * x)
    derivative = (1 - float(y) * theta_dot_x) *  (- float(y)) *  x
    return derivative

def total_error(y_set, x_set, theta, n_train):
    error = 0
    for i in range(n_train):
        error += square_loss(y_set[i], theta, x_set[i])
    return error/n_train

def update_theta(y_set, x_set, theta, n_train, eta):
    for i in range(n_train):
        derivative = square_loss_derivative(y_set[i], theta, x_set[i])
        theta -= eta * derivative
    return

step_size = 1/1000
T = 500
n_train = 10000
n_test = 100000
theta = np.zeros(100)
training_errors = []
testing_errors = []
for t in range(T):

    training_errors.append(total_error(ytrain[:n_train], Xtrain[:n_train], theta, n_train))
    testing_errors.append(total_error(ytest[:n_test], Xtest[:n_test], theta, n_test))
    update_theta(ytrain[:n_train], Xtrain[:n_train], theta, n_train, step_size)
print(training_errors)
print(testing_errors)

plt.plot(training_errors, color = "green")
plt.plot(testing_errors, color = "red")
title = "Training & Testing error for " + str(n_train) + " obs."
plt.ylabel(title)
plt.show()

n_train = 1000
n_test = 100000
T = 500
theta = np.zeros(100)
training_errors = []
testing_errors = []
for t in range(T):

    training_errors.append(total_error(ytrain[:n_train], Xtrain[:n_train], theta, n_train))
    testing_errors.append(total_error(ytest[:n_test], Xtest[:n_test], theta, n_test))
    update_theta(ytrain[:n_train], Xtrain[:n_train], theta, n_train, step_size)
print(training_errors)
print(testing_errors)

plt.plot(training_errors, color = "green")
plt.plot(testing_errors, color = "red")
title = "Training & Testing error for " + str(n_train) + " obs."
plt.ylabel(title)
plt.show()
    
    
```

```python
# For the logistic loss
import math

def logistic_loss(y, theta, x):
    theta_dot_x = np.sum(theta * x)
    return math.log(1 + math.exp(- float(y) * theta_dot_x))

def logistic_loss_derivative(y, theta, x):
    theta_dot_x = np.sum(theta * x)
    derivative = math.exp(-float(y) * theta_dot_x)/(1 + math.exp(-float(y) * theta_dot_x)) *  (- float(y)) *  x
    return derivative

def total_error(y_set, x_set, theta, n_train):
    error = 0
    for i in range(n_train):
        error += logistic_loss(y_set[i], theta, x_set[i])
    return error/n_train

def update_theta(y_set, x_set, theta, n_train, eta):
    for i in range(n_train):
        derivative = logistic_loss_derivative(y_set[i], theta, x_set[i])
        theta -= eta * derivative
    return

step_size = 1/1000
T = 500
n_train = 10000
n_test = 100000
theta = np.zeros(100)
training_errors = []
testing_errors = []
for t in range(T):

    training_errors.append(total_error(ytrain[:n_train], Xtrain[:n_train], theta, n_train))
    testing_errors.append(total_error(ytest[:n_test], Xtest[:n_test], theta, n_test))
    update_theta(ytrain[:n_train], Xtrain[:n_train], theta, n_train, step_size)
print(training_errors)
print(testing_errors)

plt.plot(training_errors, color = "green")
plt.plot(testing_errors, color = "red")
title = "Training & Testing error for " + str(n_train) + " obs."
plt.ylabel(title)
plt.show()

n_train = 1000
n_test = 100000
T = 500
theta = np.zeros(100)
training_errors = []
testing_errors = []
for t in range(T):

    training_errors.append(total_error(ytrain[:n_train], Xtrain[:n_train], theta, n_train))
    testing_errors.append(total_error(ytest[:n_test], Xtest[:n_test], theta, n_test))
    update_theta(ytrain[:n_train], Xtrain[:n_train], theta, n_train, step_size)
print(training_errors)
print(testing_errors)

plt.plot(training_errors, color = "green")
plt.plot(testing_errors, color = "red")
title = "Training & Testing error for " + str(n_train) + " obs."
plt.ylabel(title)
plt.show()
```
